```python
import os
import matplotlib.pyplot as plt
import xarray as xr

# Set data directory and filename
data_dir = "T:/SGS_FAT/Tera_Aqua_Raw_Data_L1"
files = os.listdir(data_dir)
fnames = [os.path.join(data_dir, f) for f in files]

# Open dataset with specified engine
modis_l2 = xr.open_dataset(fnames[0], engine='netcdf4')
print(modis_l2)
# Get the Cloud_Mask variable
cloud_mask = modis_l2["Cloud_Mask"]

# Determine the number of layers
num_layers = cloud_mask.shape[0]

# Plot each layer as an image
fig, axes = plt.subplots(num_layers, 1, figsize=(10, 2*num_layers))
for i in range(num_layers):
    axes[i].imshow(cloud_mask[i, :, :], origin="upper")
    axes[i].set_title(f"Cloud Mask Layer {i+1}")
    axes[i].set_xlabel("X-axis label")  # Replace with appropriate label
    axes[i].set_ylabel("Y-axis label")  # Replace with appropriate label

plt.tight_layout()
plt.show()
```
